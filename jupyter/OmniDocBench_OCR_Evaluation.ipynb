````python
import easyocr
from PIL import Image
import os


output_folder = '/content/ocr_results_easyocr'

# Initialize the EasyOCR reader with languages
reader = easyocr.Reader(['en'])#(['ch_sim', 'en'])

print("Performing OCR on sample images...")

aggregated_results = []
# Iterate through the loaded images and perform OCR
for filename in image_paths:
    print(filename)
    pred_file = os.path.join(output_folder, filename[:-4]+'.json')
    try:
        full_image_path = os.path.join(image_base_path, filename)
        img = Image.open(full_image_path)
        # Perform OCR
        ocr_results_for_image = reader.readtext(img)
        # print(ocr_results_for_image)
        detected_texts = []
        if ocr_results_for_image:
            order = 0
            for (bbox, text, prob) in ocr_results_for_image:
                detected_texts.append({
                    "category_type": "text_block",
                    "poly": [int(x) for x in flatten_concatenation(bbox)],
                    "order": order,
                    "text": text,
                    "confidence": prob,
                    "attribute": {
                        "language": "en",
                        "rotation": 0
                    }
                })
                order = order + 1

        # Create a dictionary for the current image's results
        image_data = detected_texts
        with open(pred_file, 'w', encoding='utf-8') as f:
            json.dump(image_data, f, ensure_ascii=False, indent=2)

    except FileNotFoundError:
        print(f"  Error: Image not found at {full_image_path}")
    except Exception as e:
        print(f"  Error loading {filename}: {e}")
    except Exception as e:
        print(f"Error performing OCR on image {i+1}: {e}")
````

````python
file_path = '/content/OmniDocBench/utility.py'
utility_code = """
from rtree import index
from collections import defaultdict

def calculate_iou(boxes1, boxes2):
    def union_box(boxes):
        if not boxes:
            return None
        x1 = min(b[0] for b in boxes)
        y1 = min(b[1] for b in boxes)
        x2 = max(b[2] for b in boxes)
        y2 = max(b[3] for b in boxes)
        return (x1, y1, x2, y2)

    def box_area(box):
        return max(0, box[2] - box[0]) * max(0, box[3] - box[1])

    def intersection(box1, box2):
        x1 = max(box1[0], box2[0])
        y1 = max(box1[1], box2[1])
        x2 = min(box1[2], box2[2])
        y2 = min(box1[3], box2[3])
        if x2 <= x1 or y2 <= y1:
            return 0
        return (x2 - x1) * (y2 - y1)

    union1 = union_box(boxes1)
    union2 = union_box(boxes2)

    if union1 is None or union2 is None:
        return 0.0

    inter = intersection(union1, union2)
    area1 = box_area(union1)
    area2 = box_area(union2)
    union_area = area1 + area2 - inter

    return inter / union_area if union_area > 0 else 0.0

def get_union_box(boxes):
    if not boxes:
        return None
    x1 = min(b[0] for b in boxes)
    y1 = min(b[1] for b in boxes)
    x2 = max(b[2] for b in boxes)
    y2 = max(b[3] for b in boxes)
    return (x1, y1, x2, y2)

def box_area(box):
    return max(0, box[2] - box[0]) * max(0, box[3] - box[1])

def calculate_global_score(groups, A, B):
    total_score = 0.0
    total_weight = 0.0

    for a_group, b_group in groups:
        if not a_group or not b_group:
            continue

        a_boxes = [A[i] for i in a_group]
        b_boxes = [B[i] for i in b_group]
        iou = calculate_iou(a_boxes, b_boxes)

        # Weight by size of coverage
        a_union = get_union_box(a_boxes)
        b_union = get_union_box(b_boxes)
        weight = (box_area(a_union) + box_area(b_union)) / 2

        total_score += iou * weight
        total_weight += weight

    return total_score / total_weight if total_weight > 0 else 0.0

def match_regions(A, B, iou_threshold=0.6, max_iterations=20, min_improvement=0.001):
    if not A or not B:
        return []

    # Build R-tree indices
    idx_a = index.Index()
    for i, box in enumerate(A):
        idx_a.insert(i, box)

    idx_b = index.Index()
    for i, box in enumerate(B):
        idx_b.insert(i, box)

    # Initialize groups: start with overlap-based initial grouping
    groups = []  # List of (a_set, b_set) tuples
    assigned_a = set()
    assigned_b = set()

    # Phase 1: Create initial groups based on overlaps
    for a_idx in range(len(A)):
        if a_idx in assigned_a:
            continue

        # Find overlapping B regions
        b_candidates = list(idx_b.intersection(A[a_idx]))
        b_overlap = [b for b in b_candidates if b not in assigned_b]

        if b_overlap:
            groups.append(({a_idx}, set(b_overlap)))
            assigned_a.add(a_idx)
            assigned_b.update(b_overlap)
        else:
            groups.append(({a_idx}, set()))
            assigned_a.add(a_idx)

    # Add unmatched B regions
    for b_idx in range(len(B)):
        if b_idx not in assigned_b:
            groups.append((set(), {b_idx}))
            assigned_b.add(b_idx)

    # Phase 2: Iterative refinement
    best_global_score = calculate_global_score(groups, A, B)

    for iteration in range(max_iterations):
        improved = False

        # Try to expand each group
        for group_idx, (a_group, b_group) in enumerate(groups):
            if not a_group and not b_group:
                continue

            a_boxes = [A[i] for i in a_group] if a_group else []
            b_boxes = [B[i] for i in b_group] if b_group else []

            if not a_boxes or not b_boxes:
                continue

            a_union = get_union_box(a_boxes)
            b_union = get_union_box(b_boxes)
            a_area = box_area(a_union)
            b_area = box_area(b_union)

            current_iou = calculate_iou(a_boxes, b_boxes)

            if current_iou >= iou_threshold:
                continue

            # Try expanding the smaller side
            if a_area > b_area:
                # Look for more B regions
                b_candidates = list(idx_b.intersection(a_union))

                for b_idx in b_candidates:
                    # Skip if already in this group
                    if b_idx in b_group:
                        continue

                    # Find which group owns this B region
                    owner_idx = None
                    for gi, (ag, bg) in enumerate(groups):
                        if b_idx in bg:
                            owner_idx = gi
                            break

                    if owner_idx is None:
                        continue

                    # Try moving/adding this B region to current group
                    old_a_group, old_b_group = groups[owner_idx]

                    # Create test configuration
                    test_groups = [g for g in groups]
                    test_groups[group_idx] = (a_group.copy(), b_group | {b_idx})
                    test_groups[owner_idx] = (old_a_group.copy(), old_b_group - {b_idx})

                    # Remove empty groups
                    test_groups = [(ag, bg) for ag, bg in test_groups if ag or bg]

                    test_score = calculate_global_score(test_groups, A, B)

                    if test_score > best_global_score + min_improvement:
                        groups = test_groups
                        best_global_score = test_score
                        improved = True
                        break

            else:
                # Look for more A regions
                a_candidates = list(idx_a.intersection(b_union))

                for a_idx in a_candidates:
                    if a_idx in a_group:
                        continue

                    # Find owner
                    owner_idx = None
                    for gi, (ag, bg) in enumerate(groups):
                        if a_idx in ag:
                            owner_idx = gi
                            break

                    if owner_idx is None:
                        continue

                    # Try moving this A region
                    old_a_group, old_b_group = groups[owner_idx]

                    test_groups = [g for g in groups]
                    test_groups[group_idx] = (a_group | {a_idx}, b_group.copy())
                    test_groups[owner_idx] = (old_a_group - {a_idx}, old_b_group.copy())

                    test_groups = [(ag, bg) for ag, bg in test_groups if ag or bg]

                    test_score = calculate_global_score(test_groups, A, B)

                    if test_score > best_global_score + min_improvement:
                        groups = test_groups
                        best_global_score = test_score
                        improved = True
                        break

            if improved:
                break

        # Stop if no improvement
        if not improved:
            break

    # Phase 3: Format output
    matches = []
    for a_group, b_group in groups:
        if not a_group and not b_group:
            continue

        a_boxes = [A[i] for i in a_group] if a_group else []
        b_boxes = [B[i] for i in b_group] if b_group else []

        iou = calculate_iou(a_boxes, b_boxes) if a_boxes and b_boxes else 0.0
        matches.append((sorted(a_group), sorted(b_group), iou))

    # Sort by IoU descending
    matches.sort(key=lambda x: x[2], reverse=True)

    return matches
"""
with open(file_path, 'w') as f:
    f.write(utility_code)
````
